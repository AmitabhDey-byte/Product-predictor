```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier


ingest = pd.read_csv('sample_data/amazon.csv')

```

```python
ingest = ingest.dropna()
```

```python
ingest.head()
```

```python
ingest['discounted_price']= ingest['discounted_price'].str.replace('₹','').str.replace(',','').astype(float)
ingest['actual_price']= ingest['actual_price'].str.replace('₹','').str.replace(',','').astype(float)
ingest['discount_percentage'] = ingest['discount_percentage'].str.replace('%','').astype(float)
ingest['rating_count'] = ingest['rating_count'].str.replace(',','').astype(float)

ingest.head()
```

```python
ingest['rating'] = pd.to_numeric(ingest['rating'], errors='coerce')

ingest['label']= ingest['rating'].apply(lambda x: 1 if x>=4 else 0)

ingest['label'].value_counts()
```

```python
X = ingest[['discounted_price', 'actual_price','rating_count' ]]
Y = ingest['label']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.2,stratify=Y,random_state=42)
```

```python
X.isnull().sum()
```

```python
len(ingest
    ) #ok so there are a total of 1465 rows ...there is 2 rows with Nan value..so i am dropping them..coz it will only be 0.13% of missing data
```

```python
model1 = LogisticRegression(max_iter=500)
model1.fit(X_train, Y_train)
```

```python
Y_testpred = model1.predict(X_test)

print('Test Accuracy:', accuracy_score(Y_test, Y_testpred) * 100, '%')
print('Confusion Matrix:\n', confusion_matrix(Y_test, Y_testpred))
```

```python
dt = DecisionTreeClassifier(random_state=42)

param_grid_dt = {'max_depth': [3,5,7,8,10,None],'min_samples_split': [2,5,10],'min_samples_leaf': [1,2,3,5]}
grid_dt = GridSearchCV(dt,param_grid_dt,cv=5,scoring='accuracy',n_jobs=-1)
grid_dt.fit(X_train2, Y_train2)
print("Best DT Params:", grid_dt.best_params_)
print("Best DT Score:", grid_dt.best_score_)
```
